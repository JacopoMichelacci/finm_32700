```python
import matplotlib.pyplot as plt
```

```python
n_ord_vector = [1000, 5000, 10000, 50000, 100000]
time_elapsed_vector = [367, 1801, 3368, 18446, 37846]


plt.plot(n_ord_vector, time_elapsed_vector, lw=0.8, marker="o", markersize=2)
plt.xlabel("Number of Orders")
plt.ylabel("Execution Time (microseconds)")
plt.title("HFT Order Book Performance")
plt.grid(alpha=0.6)
plt.show()
```
